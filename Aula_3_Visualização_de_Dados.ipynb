```python
df.columns = ['ano', 'senioridade', 'contrato', 'cargo',
              'salario', 'moeda', 'usd', 'residencia',
              'remoto', 'empresa', 'tamanho_empresa']

print(df.columns)
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df["senioridade"] = df["senioridade"].replace(senioridade)
df["senioridade"].value_counts()
```

```python
df.head()
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'PT': 'Meio Per√≠odo',
    'FL': 'Freelancer',
    'CT': 'Contrato'
}

df["contrato"] = df["contrato"].replace(contrato)
df["contrato"].value_counts()
```

```python
tamanho_empresa = {
    'S': 'Startup',
    'M': 'M√©dio',
    'L': 'Grande'
}

df["tamanho_empresa"] = df["tamanho_empresa"].replace(tamanho_empresa)
df["tamanho_empresa"].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    50: 'H√≠brido',
    100: 'Remoto'
}

df["remoto"] = df["remoto"].replace(remoto)
df["remoto"].value_counts()
```

```python
import numpy as np

#Cria√ß√£o da base para teste
df_salarios = pd.DataFrame({
        'nome': ['Ana', 'Jo√£o', 'Maria', 'Carol', 'Bruno'],
        'salario': [4000, np.nan, 5000, np.nan, 90000]
})

#Calcula a m√©dia salarial e substitui os nulos pela m√©dia, tamb√©m arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''Calcula a mediana e substiui os nulos pela mediana '''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame ({
    'dia': ['seg', 'ter', 'qua', 'qui', 'sex'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
display(df_temperaturas)
```

```python
df_temperaturas = pd.DataFrame ({
    'dia': ['seg', 'ter', 'qua', 'qui', 'sex'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
display(df_temperaturas)
```

```python
df_cidades = pd.DataFrame ({
    'nome': ['M√°rcio', 'Juliana', 'Felipe', 'Marcela', 'Lucas', 'Cristina'],
    'cidade': ['S√£o Paulo', np.nan, 'Flor√≠nea', np.nan, 'Ara√ßatuba', np.nan]
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('N√£o Informado')
df_cidades
```

```python
df_limpo['senioridade'].value_counts().plot(kind='bar', title='Senioridade')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd')
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = df_limpo,x = 'senioridade', y = 'usd')
plt.title('Sal√°rio m√©dio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Sal√°rio m√©dio anual em USD')
plt.show()
```

```python
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
```

```python
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = df_limpo,x = 'senioridade', y = 'usd', order = ordem)
plt.title('Sal√°rio m√©dio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Sal√°rio m√©dio anual em USD')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data = df_limpo['usd'], bins=50, kde=True)
plt.title('Distribui√ß√£o dos Sal√°rios Anuais')
plt.xlabel('Sal√°rio em USD')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x = df_limpo['usd'])
plt.title('Boxplot sal√°rio')
plt.xlabel('Sal√°rio em USD')
plt.show()
```

```python
ordem_senioridade = {'Junior', 'Pleno', 'Senior', 'Executivo'}

plt.figure(figsize=(8,5))
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade)
plt.title('Boxplot sal√°rio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Sal√°rio em USD')
plt.show()
```

```python
ordem_senioridade = {'Junior', 'Pleno', 'Senior', 'Executivo'}

plt.figure(figsize=(8,5))
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade, palette='Set2', hue='senioridade')
plt.title('Boxplot sal√°rio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Sal√°rio em USD')
plt.show()
```

```python
import plotly.express as px
```

```python
df_senioridade_media = df_limpo.groupby('senioridade')['usd'].mean().reset_index()

fig = px.bar(df_senioridade_media,
             x='senioridade',
             y='usd',
             title='Sal√°rio M√©dio por Senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'Sal√°rio M√©dio Anual (USD)'})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='propor√ß√£o dos tipos de trabalho',
          )
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='propor√ß√£o dos tipos de trabalho',
             hole = 0.5
          )
fig.update_traces(textinfo='percent+label')
fig.show()
```
